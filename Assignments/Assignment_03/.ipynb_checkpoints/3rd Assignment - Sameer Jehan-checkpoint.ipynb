# Question 1

```python
#check which columns have missing values
data.isna().sum()
```

```python
#Cast also has missing columns, so first replace all missing values in columns
```

```python
data["cast"] = data['cast'].replace(np.nan, "None")
data["country"] = data['country'].replace(np.nan, "None")
data["date_added"] = data['date_added'].replace(np.nan, "None")
data["director"] = data['director'].replace(np.nan, "None")
```

```python
data.isna().sum()
```

```python
total_cast = []
for i in range(len(data['cast'])):
    movies = data.loc[i, "cast"].split(", ")
    total_length = len(movies)
    total_cast.append(total_length)
    
print(total_cast)
```

```python
data['num_of_cast'] = total_cast
```

```python
#Changing order of columns to give a better view
data.reindex(columns = ['show_id', 'type', 'title', 'director', 'cast','num_of_cast', 'country', 'date_added','release_year', 'duration', 'description', 'genre'])
```
